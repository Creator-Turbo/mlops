**Creating labels for resized images of dogs and cats**

Cat --> 0

Dog --> 1

```python
# creaing a for loop to assign labels
filenames = os.listdir('/content/image resized/')


labels = []

for i in range(2000):

  file_name = filenames[i]
  label = file_name[0:3]

  if label == 'dog':
    labels.append(1)

  else:
    labels.append(0)
```

Converting all the resized images to numpy arrays

```python
import cv2
import glob
```

```python
image_directory = '/content/image resized/'
image_extension = ['png', 'jpg']

files = []

[files.extend(glob.glob(image_directory + '*.' + e)) for e in image_extension]

dog_cat_images = np.asarray([cv2.imread(file) for file in files])
```

```python
X = dog_cat_images
Y = np.asarray(labels)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

1600 --> training images

400 --> test images

```python
# scaling the data
X_train_scaled = X_train/255

X_test_scaled = X_test/255
```

**Building the Neural Network**

```python
import tensorflow as tf

# Define the input shape explicitly
input_shape = (224, 224, 3)

# Load the pre-trained MobileNetV2 model, excluding the top layers
pretrained_model = tf.keras.applications.MobileNetV2(input_shape=input_shape,
                                                     include_top=False,  # Exclude the top layers
                                                     weights='imagenet',  # Use ImageNet weights
                                                     pooling='avg')  # Global average pooling

num_of_classes = 2

# Create the model
inputs = tf.keras.Input(shape=input_shape)  # Explicitly define input
x = pretrained_model(inputs)  # Pass inputs through the MobileNetV2 model
outputs = tf.keras.layers.Dense(num_of_classes, activation='softmax')(x)  # Output layer

model = tf.keras.Model(inputs=inputs, outputs=outputs)

# Print model summary
model.summary()
```

```python
model.compile(
    optimizer = 'adam',
    loss = tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
    metrics = ['acc']
)
```

```python
model.fit(X_train_scaled, Y_train, epochs=5)
```

```python
score, acc = model.evaluate(X_test_scaled, Y_test)
print('Test Loss =', score)
print('Test Accuracy =', acc)
```

\**Predictive System**

```python
input_image_path = input('Path of the image to be predicted: ')

input_image = cv2.imread(input_image_path)

cv2_imshow(input_image)

input_image_resize = cv2.resize(input_image, (224,224))

input_image_scaled = input_image_resize/255

image_reshaped = np.reshape(input_image_scaled, [1,224,224,3])

input_prediction = model.predict(image_reshaped)

print(input_prediction)

input_pred_label = np.argmax(input_prediction)

print(input_pred_label)

if input_pred_label == 0:
  print('The image represents a Cat')

else:
  print('The image represents a Dog')
```
